### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular la facturación total
tienda['Total_Facturación'] = tienda['Precio'] * tienda['Cantidad de cuotas']
tienda2['Total_Facturación'] = tienda2['Precio'] * tienda2['Cantidad de cuotas']
tienda3['Total_Facturación'] = tienda3['Precio'] * tienda3['Cantidad de cuotas']
tienda4['Total_Facturación'] = tienda4['Precio'] * tienda4['Cantidad de cuotas']
```

```python
# Mostrar la facturación total por cada tienda
print("Facturación total de Tienda 1:", tienda['Total_Facturación'].sum())
print("Facturación total de Tienda 2:", tienda2['Total_Facturación'].sum())
print("Facturación total de Tienda 3:", tienda3['Total_Facturación'].sum())
print("Facturación total de Tienda 4:", tienda4['Total_Facturación'].sum())
```

# 2. Ventas por categoría

```python
# Calcular ventas por categorías por cada tienda
sales_by_category_tienda1 = tienda.groupby('Categoría del Producto')['Total_Facturación'].sum()
sales_by_category_tienda2 = tienda2.groupby('Categoría del Producto')['Total_Facturación'].sum()
sales_by_category_tienda3 = tienda3.groupby('Categoría del Producto')['Total_Facturación'].sum()
sales_by_category_tienda4 = tienda4.groupby('Categoría del Producto')['Total_Facturación'].sum()
```

```python
# Mostrar ventas por categorías por cada tienda
print("Ventas por categoría - Tienda 1:")
display(sales_by_category_tienda1)
print("\nVentas por categoría - Tienda 2:")
display(sales_by_category_tienda2)
print("\nVentas por categoría - Tienda 3:")
display(sales_by_category_tienda3)
print("\nVentas por categoría - Tienda 4:")
display(sales_by_category_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación pronedio de cada tienda
average_rating_tienda1 = tienda['Calificación'].mean()
average_rating_tienda2 = tienda2['Calificación'].mean()
average_rating_tienda3 = tienda3['Calificación'].mean()
average_rating_tienda4 = tienda4['Calificación'].mean()
```

```python
# Mostrar la calificación pronedio de cada tienda
print("Calificación promedio - Tienda 1:", average_rating_tienda1)
print("Calificación promedio - Tienda 2:", average_rating_tienda2)
print("Calificación promedio - Tienda 3:", average_rating_tienda3)
print("Calificación promedio - Tienda 4:", average_rating_tienda4)
```

# 4. Productos más y menos vendidos

```python
# Calcular productos más y menos vendidos por tienda
sales_by_product_tienda1 = tienda.groupby('Producto')['Total_Facturación'].sum()
sales_by_product_tienda2 = tienda2.groupby('Producto')['Total_Facturación'].sum()
sales_by_product_tienda3 = tienda3.groupby('Producto')['Total_Facturación'].sum()
sales_by_product_tienda4 = tienda4.groupby('Producto')['Total_Facturación'].sum()
```

```python
# Mostrar los productos más y menos vendidos por tienda
print("Productos más y menos vendidos - Tienda 1:")
print("Más vendidos:")
display(sales_by_product_tienda1.nlargest(5))
print("\nMenos vendidos:")
display(sales_by_product_tienda1.nsmallest(5))

print("\nProductos más y menos vendidos - Tienda 2:")
print("Más vendidos:")
display(sales_by_product_tienda2.nlargest(5))
print("\nMenos vendidos:")
display(sales_by_product_tienda2.nsmallest(5))

print("\nProductos más y menos vendidos - Tienda 3:")
print("Más vendidos:")
display(sales_by_product_tienda3.nlargest(5))
print("\nMenos vendidos:")
display(sales_by_product_tienda3.nsmallest(5))

print("\nProductos más y menos vendidos - Tienda 4:")
print("Más vendidos:")
display(sales_by_product_tienda4.nlargest(5))
print("\nMenos vendidos:")
display(sales_by_product_tienda4.nsmallest(5))
```

# 5. Envío promedio por tienda

```python
# Calcular envío promedio por tienda
average_shipping_tienda1 = tienda['Costo de envío'].mean()
average_shipping_tienda2 = tienda2['Costo de envío'].mean()
average_shipping_tienda3 = tienda3['Costo de envío'].mean()
average_shipping_tienda4 = tienda4['Costo de envío'].mean()
```

```python
# Mostrar el envío promedio por tienda
print("Envío promedio - Tienda 1:", average_shipping_tienda1)
print("Envío promedio - Tienda 2:", average_shipping_tienda2)
print("Envío promedio - Tienda 3:", average_shipping_tienda3)
print("Envío promedio - Tienda 4:", average_shipping_tienda4)
```
